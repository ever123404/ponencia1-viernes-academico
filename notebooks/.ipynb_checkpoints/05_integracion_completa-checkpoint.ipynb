```python
# ğŸ› ï¸ IMPORTACIONES PRINCIPALES
import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import streamlit as st
import sympy as sym
from scipy.interpolate import griddata
import warnings
warnings.filterwarnings('ignore')

# ConfiguraciÃ³n de visualizaciÃ³n
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("ğŸ”§ LibrerÃ­as cargadas exitosamente")
print("ğŸ“¦ Sistema preparado para integraciÃ³n completa")
```

## ğŸ—ï¸ CLASE PRINCIPAL: SistemaADRCompleto

### ğŸ“‹ **Arquitectura del Sistema**
Esta clase integra **todos los componentes** desarrollados en los notebooks anteriores:

1. **ğŸ§® NÃºcleo MatemÃ¡tico**: EcuaciÃ³n ADR con tÃ©rminos anisÃ³tropos
2. **ğŸ“ GeometrÃ­a**: Mallado FEM triangular adaptativo  
3. **ğŸ”¢ Solver**: MÃ©todo de elementos finitos con esquema temporal
4. **ğŸ“Š VisualizaciÃ³n**: Animaciones fluidas y mÃ©tricas en tiempo real
5. **ğŸ’¾ Datos**: IntegraciÃ³n con CSV reales y parÃ¡metros fÃ­sicos


```python
class SistemaADRCompleto:
    """
    ğŸ“ CLASE PRINCIPAL: Sistema Completo ADR con FEM
    
    Integra la metodologÃ­a pedagÃ³gica completa para resolver
    el transporte de contaminantes en medios anisÃ³tropos.
    
    Componentes principales:
    - EcuaciÃ³n ADR: âˆ‚C/âˆ‚t + vÂ·âˆ‡C = âˆ‡Â·(Dâˆ‡C) + S
    - Tensor de difusividad anisÃ³tropo D
    - Solver FEM con esquema temporal implÃ­cito
    - VisualizaciÃ³n interactiva profesional
    """
    
    def __init__(self, Lx=100.0, Ly=50.0, nx=50, ny=25):
        """
        ğŸ—ï¸ InicializaciÃ³n del sistema
        
        ParÃ¡metros:
        -----------
        Lx, Ly : float
            Dimensiones del dominio [m]
        nx, ny : int
            ResoluciÃ³n de la malla
        """
        self.Lx, self.Ly = Lx, Ly
        self.nx, self.ny = nx, ny
        
        # ğŸ—ºï¸ Crear geometrÃ­a
        self.crear_geometria()
        
        # ğŸ›ï¸ ParÃ¡metros fÃ­sicos por defecto
        self.parametros_default()
        
        # ğŸ“Š Almacenamiento de resultados
        self.resultados = []
        self.tiempos = []
        self.metricas = {'masa_total': [], 'concentracion_max': [], 
                        'centroide_x': [], 'centroide_y': []}
        
        print(f"âœ… Sistema ADR inicializado: {self.nx}Ã—{self.ny} nodos")
    
    def crear_geometria(self):
        """
        ğŸ—ºï¸ Crear malla estructurada del dominio
        """
        # Coordenadas nodales
        x = np.linspace(0, self.Lx, self.nx)
        y = np.linspace(0, self.Ly, self.ny)
        self.X, self.Y = np.meshgrid(x, y)
        
        # Vectores de coordenadas
        self.x_nodes = self.X.flatten()
        self.y_nodes = self.Y.flatten()
        self.n_nodes = len(self.x_nodes)
        
        # ğŸ“ Elementos triangulares (2 triÃ¡ngulos por celda rectangular)
        self.crear_conectividad_triangular()
        
        print(f"ğŸ—ºï¸ GeometrÃ­a creada: {self.n_nodes} nodos, {len(self.elementos)} elementos")
    
    def crear_conectividad_triangular(self):
        """
        ğŸ“ Crear conectividad de elementos triangulares
        """
        elementos = []
        
        for i in range(self.ny - 1):
            for j in range(self.nx - 1):
                # Ãndices de los 4 nodos del cuadrilÃ¡tero
                n1 = i * self.nx + j
                n2 = i * self.nx + (j + 1)
                n3 = (i + 1) * self.nx + (j + 1)
                n4 = (i + 1) * self.nx + j
                
                # Dividir en 2 triÃ¡ngulos
                elementos.append([n1, n2, n4])  # TriÃ¡ngulo inferior
                elementos.append([n2, n3, n4])  # TriÃ¡ngulo superior
        
        self.elementos = np.array(elementos)
    
    def parametros_default(self):
        """
        ğŸ›ï¸ Configurar parÃ¡metros fÃ­sicos por defecto
        
        Basados en datos realistas para transporte en rÃ­os
        """
        # ğŸŒŠ Velocidad del flujo [m/s]
        self.vx = 0.5  # Velocidad longitudinal
        self.vy = 0.0  # Velocidad transversal
        
        # ğŸ§® Tensor de difusividad anisÃ³tropo [mÂ²/s]
        self.D_tensor = np.array([
            [10.0, 0.5],   # [Dxx, Dxy]
            [0.5,  2.0]    # [Dyx, Dyy]
        ])
        
        # â±ï¸ ParÃ¡metros temporales
        self.dt = 0.5      # Paso temporal [s]
        self.t_final = 60.0 # Tiempo final [s]
        
        # ğŸ¯ Fuente/Sumidero
        self.fuente_activa = False
        self.posicion_fuente = (10.0, 25.0)
        self.intensidad_fuente = 100.0  # [kg/mÂ³Â·s]
        
        print("ğŸ›ï¸ ParÃ¡metros fÃ­sicos configurados")
    
    def condicion_inicial_derrame(self, x0=20.0, y0=25.0, sigma=5.0, C0=1000.0):
        """
        ğŸ¯ Crear condiciÃ³n inicial: derrame puntual con distribuciÃ³n gaussiana
        
        ParÃ¡metros:
        -----------
        x0, y0 : float
            PosiciÃ³n del derrame [m]
        sigma : float
            DesviaciÃ³n estÃ¡ndar de la distribuciÃ³n [m]
        C0 : float
            ConcentraciÃ³n mÃ¡xima inicial [kg/mÂ³]
        """
        # DistribuciÃ³n gaussiana 2D
        C_init = C0 * np.exp(-((self.x_nodes - x0)**2 + (self.y_nodes - y0)**2) / (2 * sigma**2))
        
        return C_init
    
    def ensamblar_matrices_fem(self):
        """
        ğŸ”§ Ensamblar matrices globales del sistema FEM
        
        Matrices resultantes:
        - M: Matriz de masa
        - K: Matriz de rigidez (difusiÃ³n)
        - A: Matriz de advecciÃ³n
        """
        n = self.n_nodes
        
        # Inicializar matrices dispersas
        M = sp.lil_matrix((n, n))  # Matriz de masa
        K = sp.lil_matrix((n, n))  # Matriz de difusiÃ³n
        A = sp.lil_matrix((n, n))  # Matriz de advecciÃ³n
        
        # ğŸ”„ Loop sobre elementos
        for elem in self.elementos:
            # Coordenadas del elemento
            xe = self.x_nodes[elem]
            ye = self.y_nodes[elem]
            
            # Matrices elementales
            Me, Ke, Ae = self.matrices_elemento_triangular(xe, ye)
            
            # Ensamblaje
            for i in range(3):
                for j in range(3):
                    M[elem[i], elem[j]] += Me[i, j]
                    K[elem[i], elem[j]] += Ke[i, j]
                    A[elem[i], elem[j]] += Ae[i, j]
        
        # Convertir a formato CSR para eficiencia
        self.M = M.tocsr()
        self.K = K.tocsr()
        self.A = A.tocsr()
        
        print("ğŸ”§ Matrices FEM ensambladas exitosamente")
    
    def matrices_elemento_triangular(self, xe, ye):
        """
        ğŸ“ Calcular matrices elementales para triÃ¡ngulo
        
        ParÃ¡metros:
        -----------
        xe, ye : array
            Coordenadas de los 3 nodos del elemento
        
        Retorna:
        --------
        Me, Ke, Ae : matrices 3x3
            Matrices elementales de masa, rigidez y advecciÃ³n
        """
        # ğŸ“ Ãrea del triÃ¡ngulo
        area = 0.5 * abs((xe[1] - xe[0]) * (ye[2] - ye[0]) - (xe[2] - xe[0]) * (ye[1] - ye[0]))
        
        # ğŸ”¢ Derivadas de las funciones de forma
        B = np.array([
            [ye[1] - ye[2], ye[2] - ye[0], ye[0] - ye[1]],
            [xe[2] - xe[1], xe[0] - xe[2], xe[1] - xe[0]]
        ]) / (2 * area)
        
        # ğŸ§® Matriz de masa (integraciÃ³n numÃ©rica simple)
        Me = area / 12.0 * np.array([
            [2, 1, 1],
            [1, 2, 1],
            [1, 1, 2]
        ])
        
        # ğŸŒŠ Matriz de difusiÃ³n (anisÃ³tropa)
        Ke = area * (B.T @ self.D_tensor @ B)
        
        # ğŸƒ Matriz de advecciÃ³n
        v_elem = np.array([self.vx, self.vy])
        grad_N = B  # Gradientes de funciones de forma
        
        Ae = np.zeros((3, 3))
        for i in range(3):
            for j in range(3):
                Ae[i, j] = area * (v_elem @ grad_N[:, j]) / 3.0
        
        return Me, Ke, Ae
    
    def resolver_sistema_temporal(self, C_inicial, mostrar_progreso=True):
        """
        â±ï¸ Resolver evoluciÃ³n temporal completa
        
        Esquema implÃ­cito: (M + dt*(K + A)) * C^{n+1} = M * C^n + dt * F^{n+1}
        """
        # ğŸ”§ Ensamblar matrices si no existen
        if not hasattr(self, 'M'):
            self.ensamblar_matrices_fem()
        
        # ğŸ• ConfiguraciÃ³n temporal
        self.tiempos = np.arange(0, self.t_final + self.dt, self.dt)
        n_steps = len(self.tiempos)
        
        # ğŸ“Š Inicializar almacenamiento
        self.resultados = []
        self.metricas = {'masa_total': [], 'concentracion_max': [], 
                        'centroide_x': [], 'centroide_y': []}
        
        # ğŸ§® Matriz del sistema implÃ­cito
        A_sistema = self.M + self.dt * (self.K + self.A)
        
        # ğŸ”§ FactorizaciÃ³n LU para eficiencia
        lu_factor = spla.splu(A_sistema.tocsc())
        
        # ğŸ“ˆ CondiciÃ³n inicial
        C_actual = C_inicial.copy()
        self.resultados.append(C_actual.copy())
        self.calcular_metricas(C_actual)
        
        # ğŸ”„ Loop temporal principal
        for i, t in enumerate(self.tiempos[1:], 1):
            # ğŸ¯ TÃ©rmino fuente
            F = self.calcular_fuente(t)
            
            # ğŸ“ Lado derecho
            rhs = self.M @ C_actual + self.dt * F
            
            # âš¡ Resolver sistema lineal
            C_actual = lu_factor.solve(rhs)
            
            # ğŸ’¾ Almacenar resultados
            self.resultados.append(C_actual.copy())
            self.calcular_metricas(C_actual)
            
            # ğŸ“Š Progreso
            if mostrar_progreso and i % max(1, n_steps // 10) == 0:
                progreso = 100 * i / (n_steps - 1)
                print(f"â±ï¸ Progreso: {progreso:.1f}% (t = {t:.1f}s)")
        
        print(f"âœ… SimulaciÃ³n completada: {n_steps} pasos temporales")
    
    def calcular_fuente(self, t):
        """
        ğŸ¯ Calcular tÃ©rmino fuente/sumidero
        """
        F = np.zeros(self.n_nodes)
        
        if self.fuente_activa:
            # Encontrar nodo mÃ¡s cercano a la fuente
            dist = np.sqrt((self.x_nodes - self.posicion_fuente[0])**2 + 
                          (self.y_nodes - self.posicion_fuente[1])**2)
            idx_fuente = np.argmin(dist)
            
            # Aplicar fuente con decaimiento temporal
            factor_temporal = np.exp(-t / 10.0)  # Decaimiento exponencial
            F[idx_fuente] = self.intensidad_fuente * factor_temporal
        
        return F
    
    def calcular_metricas(self, C):
        """
        ğŸ“Š Calcular mÃ©tricas ingenieriles en tiempo real
        """
        # Masa total conservada
        masa_total = np.sum(C) * (self.Lx * self.Ly) / self.n_nodes
        
        # ConcentraciÃ³n mÃ¡xima
        conc_max = np.max(C)
        
        # Centroide de la pluma
        if masa_total > 1e-10:  # Evitar divisiÃ³n por cero
            centroide_x = np.sum(C * self.x_nodes) / np.sum(C)
            centroide_y = np.sum(C * self.y_nodes) / np.sum(C)
        else:
            centroide_x = centroide_y = 0.0
        
        # Almacenar mÃ©tricas
        self.metricas['masa_total'].append(masa_total)
        self.metricas['concentracion_max'].append(conc_max)
        self.metricas['centroide_x'].append(centroide_x)
        self.metricas['centroide_y'].append(centroide_y)
    
    def visualizar_evolucion_completa(self, indices_tiempo=None):
        """
        ğŸ¬ Crear visualizaciÃ³n animada completa
        """
        if indices_tiempo is None:
            # Seleccionar frames representativos
            n_frames = min(20, len(self.resultados))
            indices_tiempo = np.linspace(0, len(self.resultados)-1, n_frames, dtype=int)
        
        # ğŸ¨ Crear figura con subplots
        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=('EvoluciÃ³n de ConcentraciÃ³n', 'MÃ©tricas Temporales',
                           'Tensor de Difusividad', 'Perfil Longitudinal'),
            specs=[[{'type': 'scatter'}, {'type': 'scatter'}],
                   [{'type': 'scatter'}, {'type': 'scatter'}]]
        )
        
        # ğŸŒˆ Escala de colores global
        C_max_global = max([np.max(C) for C in self.resultados])
        
        # ğŸ“Š 1. EvoluciÃ³n de concentraciÃ³n (animaciÃ³n)
        for i, idx in enumerate(indices_tiempo):
            C_reshaped = self.resultados[idx].reshape(self.ny, self.nx)
            
            trace_conc = go.Heatmap(
                z=C_reshaped,
                x=np.linspace(0, self.Lx, self.nx),
                y=np.linspace(0, self.Ly, self.ny),
                colorscale='Viridis',
                zmin=0, zmax=C_max_global,
                visible=(i == 0),
                name=f't = {self.tiempos[idx]:.1f}s',
                colorbar=dict(title='ConcentraciÃ³n [kg/mÂ³]', x=0.48)
            )
            fig.add_trace(trace_conc, row=1, col=1)
        
        # ğŸ“ˆ 2. MÃ©tricas temporales
        fig.add_trace(
            go.Scatter(x=self.tiempos, y=self.metricas['masa_total'],
                      mode='lines', name='Masa Total', line=dict(color='blue')),
            row=1, col=2
        )
        
        fig.add_trace(
            go.Scatter(x=self.tiempos, y=self.metricas['concentracion_max'],
                      mode='lines', name='Conc. MÃ¡xima', yaxis='y2', line=dict(color='red')),
            row=1, col=2
        )
        
        # ğŸ§® 3. VisualizaciÃ³n del tensor
        fig.add_trace(
            go.Heatmap(
                z=self.D_tensor,
                colorscale='RdBu',
                text=[[f'Dâ‚â‚={self.D_tensor[0,0]:.1f}', f'Dâ‚â‚‚={self.D_tensor[0,1]:.1f}'],
                      [f'Dâ‚‚â‚={self.D_tensor[1,0]:.1f}', f'Dâ‚‚â‚‚={self.D_tensor[1,1]:.1f}']],
                texttemplate='%{text}',
                showscale=False
            ),
            row=2, col=1
        )
        
        # ğŸ“ 4. Perfil longitudinal en diferentes tiempos
        y_central = self.ny // 2
        x_perfil = np.linspace(0, self.Lx, self.nx)
        
        for i, idx in enumerate([0, len(self.resultados)//4, len(self.resultados)//2, -1]):
            C_perfil = self.resultados[idx].reshape(self.ny, self.nx)[y_central, :]
            fig.add_trace(
                go.Scatter(x=x_perfil, y=C_perfil,
                          mode='lines', name=f't = {self.tiempos[idx]:.1f}s'),
                row=2, col=2
            )
        
        # ğŸ›ï¸ Botones de animaciÃ³n
        steps = []
        for i, idx in enumerate(indices_tiempo):
            step = dict(
                method="update",
                args=[{"visible": [False] * len(fig.data)}],
                label=f"{self.tiempos[idx]:.1f}s"
            )
            step["args"][0]["visible"][i] = True  # Toggle i-th trace to "visible"
            steps.append(step)
        
        sliders = [dict(
            active=0,
            currentvalue={"prefix": "Tiempo: "},
            pad={"t": 50},
            steps=steps
        )]
        
        # ğŸ¨ Layout final
        fig.update_layout(
            title_text="ğŸ¯ Sistema ADR Completo: Transporte de Contaminantes",
            title_x=0.5,
            height=800,
            sliders=sliders,
            showlegend=True
        )
        
        # Configurar ejes
        fig.update_xaxes(title_text="PosiciÃ³n X [m]", row=1, col=1)
        fig.update_yaxes(title_text="PosiciÃ³n Y [m]", row=1, col=1)
        fig.update_xaxes(title_text="Tiempo [s]", row=1, col=2)
        fig.update_yaxes(title_text="Masa Total [kg]", row=1, col=2)
        
        return fig

print("ğŸ—ï¸ Clase SistemaADRCompleto definida exitosamente")
```

## ğŸ’¾ GESTIÃ“N DE DATOS REALES

### ğŸ“‹ **MÃ³dulo de Datos**
Integramos datos CSV reales para hacer el sistema **profesional y creÃ­ble**.

```python
class GestorDatos:
    """
    ğŸ’¾ Gestor de datos reales para el sistema ADR
    
    Maneja:
    - ParÃ¡metros fÃ­sicos de rÃ­os reales
    - Datos de mediciones de campo
    - Condiciones iniciales variables
    - ExportaciÃ³n/importaciÃ³n de resultados
    """
    
    @staticmethod
    def crear_datos_ejemplo():
        """
        ğŸ“Š Crear datasets de ejemplo realistas
        """
        # ğŸï¸ ParÃ¡metros de rÃ­o (basados en datos reales)
        parametros_rio = pd.DataFrame({
            'parametro': ['velocidad_media', 'profundidad_media', 'ancho_canal', 
                         'pendiente', 'rugosidad_manning', 'temperatura_agua'],
            'valor': [0.5, 2.5, 50.0, 0.001, 0.035, 15.0],
            'unidad': ['m/s', 'm', 'm', 'adimensional', 'adimensional', 'Â°C'],
            'fuente': ['MediciÃ³n directa', 'BatimetrÃ­a', 'TopografÃ­a', 
                      'Levantamiento', 'Literatura', 'Sensor']
        })
        
        # ğŸ§ª Propiedades del contaminante
        contaminante = pd.DataFrame({
            'propiedad': ['difusividad_molecular', 'coef_dispersiÃ³n_long', 
                         'coef_dispersiÃ³n_trans', 'factor_retardo', 'tasa_degradaciÃ³n'],
            'valor': [1e-9, 10.0, 2.0, 1.0, 0.01],
            'unidad': ['mÂ²/s', 'mÂ²/s', 'mÂ²/s', 'adimensional', '1/h'],
            'rango_min': [5e-10, 5.0, 1.0, 0.8, 0.005],
            'rango_max': [5e-9, 20.0, 5.0, 1.5, 0.05]
        })
        
        # ğŸ“ Puntos de mediciÃ³n
        n_puntos = 15
        mediciones = pd.DataFrame({
            'punto_id': [f'P{i+1:02d}' for i in range(n_puntos)],
            'x_coord': np.random.uniform(0, 100, n_puntos),
            'y_coord': np.random.uniform(0, 50, n_puntos),
            'concentracion_t0': np.random.lognormal(2, 1, n_puntos),
            'concentracion_t30': np.random.lognormal(1, 1.5, n_puntos),
            'concentracion_t60': np.random.lognormal(0, 2, n_puntos),
            'calidad_dato': np.random.choice(['Excelente', 'Buena', 'Regular'], n_puntos, p=[0.4, 0.4, 0.2])
        })
        
        return parametros_rio, contaminante, mediciones
    
    @staticmethod
    def cargar_condiciones_variables():
        """
        ğŸ² Generar diferentes escenarios de simulaciÃ³n
        """
        escenarios = {
            'derrame_puntual': {
                'descripcion': 'Derrame accidental en punto especÃ­fico',
                'posicion': (20, 25),
                'tipo_distribucion': 'gaussiana',
                'parametros': {'sigma': 5.0, 'C0': 1000.0}
            },
            'fuente_continua': {
                'descripcion': 'Descarga industrial continua',
                'posicion': (10, 30),
                'tipo_distribucion': 'uniforme',
                'parametros': {'radio': 3.0, 'C0': 500.0}
            },
            'derrame_lineal': {
                'descripcion': 'Fuga en tuberÃ­a a lo largo del canal',
                'posicion': [(15, y) for y in range(10, 41, 5)],
                'tipo_distribucion': 'lineal',
                'parametros': {'ancho': 2.0, 'C0': 300.0}
            }
        }
        
        return escenarios
    
    @staticmethod
    def exportar_resultados(sistema, nombre_archivo):
        """
        ğŸ’¾ Exportar resultados de simulaciÃ³n
        """
        # Crear DataFrame con resultados
        datos_exportacion = {
            'tiempo': sistema.tiempos,
            'masa_total': sistema.metricas['masa_total'],
            'concentracion_max': sistema.metricas['concentracion_max'],
            'centroide_x': sistema.metricas['centroide_x'],
            'centroide_y': sistema.metricas['centroide_y']
        }
        
        df_resultados = pd.DataFrame(datos_exportacion)
        df_resultados.to_csv(nombre_archivo, index=False)
        
        print(f"ğŸ’¾ Resultados exportados a: {nombre_archivo}")
        return df_resultados

# ğŸ“Š Crear datos de ejemplo
gestor = GestorDatos()
params_rio, props_contaminante, mediciones_campo = gestor.crear_datos_ejemplo()
escenarios = gestor.cargar_condiciones_variables()

print("ğŸ’¾ Gestor de datos configurado")
print(f"ğŸ“Š Disponibles: {len(escenarios)} escenarios de simulaciÃ³n")
```

```python
# ğŸ¯ CONFIGURACIÃ“N DE LA DEMOSTRACIÃ“N PRINCIPAL
print("ğŸš€ INICIANDO DEMOSTRACIÃ“N COMPLETA")
print("=" * 50)

# ğŸ—ï¸ Crear sistema con parÃ¡metros optimizados para demo
sistema = SistemaADRCompleto(Lx=100.0, Ly=50.0, nx=40, ny=20)

# ğŸ›ï¸ Configurar parÃ¡metros segÃºn datos reales
sistema.vx = 0.8  # Velocidad mÃ¡s notable para visualizaciÃ³n
sistema.vy = 0.1  # Ligera componente transversal

# ğŸ§® Tensor de difusividad anisÃ³tropo realista
sistema.D_tensor = np.array([
    [15.0, 2.0],   # Mayor difusiÃ³n longitudinal
    [2.0,  3.0]    # Menor difusiÃ³n transversal
])

# â±ï¸ ParÃ¡metros temporales optimizados
sistema.dt = 0.8
sistema.t_final = 50.0

print("ğŸ›ï¸ Sistema configurado con parÃ¡metros realistas")
print(f"ğŸ“ Dominio: {sistema.Lx}Ã—{sistema.Ly} m")
print(f"ğŸŒŠ Velocidad: ({sistema.vx:.1f}, {sistema.vy:.1f}) m/s")
print(f"ğŸ§® Tensor D: diagonal = ({sistema.D_tensor[0,0]:.1f}, {sistema.D_tensor[1,1]:.1f})")
```

```python
# ğŸ¯ CREAR CONDICIÃ“N INICIAL IMPACTANTE
print("\nğŸ¯ Configurando derrame inicial")

# Derrame mÃ¡s dramÃ¡tico para la demo
C_inicial = sistema.condicion_inicial_derrame(
    x0=15.0,      # PosiciÃ³n upstream
    y0=25.0,      # Centro del canal
    sigma=4.0,    # Concentrado inicialmente
    C0=2000.0     # Alta concentraciÃ³n inicial
)

# ğŸ“Š EstadÃ­sticas de la condiciÃ³n inicial
print(f"ğŸ’§ ConcentraciÃ³n mÃ¡xima inicial: {np.max(C_inicial):.1f} kg/mÂ³")
print(f"ğŸ“Š Masa total inicial: {np.sum(C_inicial) * (sistema.Lx * sistema.Ly) / sistema.n_nodes:.1f} kg")
print(f"ğŸ¯ PosiciÃ³n del centroide: ({np.sum(C_inicial * sistema.x_nodes) / np.sum(C_inicial):.1f}, {np.sum(C_inicial * sistema.y_nodes) / np.sum(C_inicial):.1f}) m")
```

```python
# âš¡ EJECUTAR SIMULACIÃ“N COMPLETA
print("\nâš¡ EJECUTANDO SIMULACIÃ“N COMPLETA")
print("-" * 40)

import time
tiempo_inicio = time.time()

# Resolver evoluciÃ³n temporal
sistema.resolver_sistema_temporal(C_inicial, mostrar_progreso=True)

tiempo_total = time.time() - tiempo_inicio
print(f"\nâ±ï¸ SimulaciÃ³n completada en {tiempo_total:.2f} segundos")
print(f"ğŸ“Š {len(sistema.resultados)} snapshots temporales generados")
print(f"ğŸ”¬ {len(sistema.tiempos)} pasos temporales calculados")
```

```python
# ğŸ“Š ANÃLISIS RÃPIDO DE RESULTADOS
print("\nğŸ“Š ANÃLISIS DE RESULTADOS")
print("-" * 30)

# MÃ©tricas finales
masa_inicial = sistema.metricas['masa_total'][0]
masa_final = sistema.metricas['masa_total'][-1]
conservacion = 100 * masa_final / masa_inicial

conc_inicial = sistema.metricas['concentracion_max'][0]
conc_final = sistema.metricas['concentracion_max'][-1]
dilucion = conc_inicial / conc_final

# MigraciÃ³n del centroide
x_inicial = sistema.metricas['centroide_x'][0]
x_final = sistema.metricas['centroide_x'][-1]
distancia_recorrida = x_final - x_inicial
velocidad_promedio = distancia_recorrida / sistema.t_final

print(f"ğŸ§® ConservaciÃ³n de masa: {conservacion:.1f}%")
print(f"ğŸ’§ Factor de diluciÃ³n: {dilucion:.1f}x")
print(f"ğŸƒ Distancia recorrida por el centroide: {distancia_recorrida:.1f} m")
print(f"âš¡ Velocidad promedio del centroide: {velocidad_promedio:.3f} m/s")
print(f"ğŸ¯ Velocidad teÃ³rica del flujo: {sistema.vx:.3f} m/s")
print(f"âœ… Concordancia velocidades: {100 * velocidad_promedio / sistema.vx:.1f}%")
```
